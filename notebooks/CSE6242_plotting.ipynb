```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
with h5py.File(r'C:\Users\willd\Documents\Georgia Tech\CSE6424\Project\nearest.hdf5', 'r') as f:
    nodes_x = f['nodes_x'][:]
    nodes_y = f['nodes_y'][:]
    food_travel_time = f['closest_food_nodes_travel_time'][:]
    physical_travel_time = f['closest_physical_nodes_travel_time'][:]
    transport_travel_time = f['closest_transport_nodes_travel_time'][:]
    education_travel_time = f['closest_education_nodes_travel_time'][:]
    worship_travel_time = f['closest_worship_nodes_travel_time'][:]
    print(f.keys())
smaller_bounds = (nodes_x > -74.4) & (nodes_x < -73.6) & (nodes_y < 41)
nodes_x = nodes_x[smaller_bounds]
nodes_y = nodes_y[smaller_bounds]
food_travel_time = food_travel_time[smaller_bounds]
physical_travel_time = physical_travel_time[smaller_bounds]
transport_travel_time = transport_travel_time[smaller_bounds]
education_travel_time = education_travel_time[smaller_bounds]
worship_travel_time = worship_travel_time[smaller_bounds]
```

```python
min_food = np.quantile(food_travel_time, 0.1)
max_food = np.quantile(food_travel_time, 0.9)
min_physical = np.quantile(physical_travel_time, 0.1)
max_physical = np.quantile(physical_travel_time, 0.9)
min_transport = np.quantile(transport_travel_time, 0.1)
max_transport = np.quantile(transport_travel_time, 0.9)
min_education = np.quantile(education_travel_time, 0.1)
max_education = np.quantile(education_travel_time, 0.9)
min_worship = np.quantile(worship_travel_time, 0.1)
max_worship = np.quantile(worship_travel_time, 0.9)
```

```python
def normalize(x, min_, max_):
    return (x - min_)/(max_ - min_)
```

```python
plt.scatter(nodes_x, nodes_y, c=[cm.coolwarm(normalize(x, min_food, max_food)) for x in food_travel_time], s=3)
plt.title('Food travel time')
plt.savefig(r'C:\Users\willd\Documents\Georgia Tech\CSE6424\Project\food_travel_time.png', dpi=300)
```

```python
plt.scatter(nodes_x, nodes_y, c=[cm.coolwarm(normalize(x, min_physical, max_physical)) for x in physical_travel_time], s=3)
plt.title('Physical activity travel time')
plt.savefig(r'C:\Users\willd\Documents\Georgia Tech\CSE6424\Project\physical_travel_time.png', dpi=300)
```
